```python
# Dependencies
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import datetime
```

# PCR Test Counts 

```python
# path to CSV file
PCR_file = 'Resources/COVID-19_Diagnostic_Laboratory_Testing__PCR_Testing__Time_Series.csv'
PCR_file
```

```python
# Convert CSV in pandas dataframe
PCR_df = pd.read_csv(PCR_file, encoding="ISO-8859-1")
PCR_df.head(5)
```

# Test Positive PCR 

```python
# Select test positive PCR results
Positive_PCR_df = PCR_df.loc[(PCR_df["overall_outcome"]=="Positive")]
Positive_PCR_df.head(10)
```

# State Selection 

```python
# Select States (CA, WA, NY, OH, FL, LA, MD, GA) for Project
State_Positive_PCR_df = Positive_PCR_df.loc[(Positive_PCR_df["state"]=="CA") | (Positive_PCR_df["state"]=="WA") | (Positive_PCR_df["state"]=="NY") | (Positive_PCR_df["state"]=="OH") | (Positive_PCR_df["state"]=="FL") | (Positive_PCR_df["state"]=="LA") | (Positive_PCR_df["state"]=="MD") | (Positive_PCR_df["state"]=="GA")]
State_Positive_PCR_df.head(10)
```

```python
# Select requied field for Data analysis
Date_State_Positive_PCR_df = State_Positive_PCR_df.loc[:, ["state", "date", "new_results_reported", "total_results_reported", "geocoded_state"]]
Date_State_Positive_PCR_df.head()
```

```python
# clean "geocoded_state" in to geocode
Date_State_Positive_PCR_df['geocode'] = Date_State_Positive_PCR_df['geocoded_state'].str.replace('POINT', '')
```

```python
Date_State_Positive_PCR_df.head(5)
```

```python
# clean "geocode"
Date_State_Positive_PCR_df['geocode'] = Date_State_Positive_PCR_df['geocode'].str.replace('(', '').str.replace(')', '')
Date_State_Positive_PCR_df.head(5)
```

```python
Date_State_Positive_PCR_df['geocode'] = Date_State_Positive_PCR_df['geocode'].str.split(' ')
Date_State_Positive_PCR_df.head(5)
```

```python
# Split geocode in to Longitude and Latitude
Date_State_Positive_PCR_df['Longitude'] = Date_State_Positive_PCR_df['geocode'].str.get(1)
Date_State_Positive_PCR_df['Latitude'] = Date_State_Positive_PCR_df['geocode'].str.get(2)
Date_State_Positive_PCR_df.head(5)
```

```python
# Final table creation
Date_State_Positive_PCR_df = Date_State_Positive_PCR_df.loc[:, ["state", "date", "new_results_reported", "total_results_reported", "Longitude", "Latitude"]]
Date_State_Positive_PCR_df.head()
```

```python
# Sort final table 
Date_State_Positive_PCR_df = Date_State_Positive_PCR_df.sort_values (by =["date"], axis = 0, ascending=[True])
Date_State_Positive_PCR_df
```

```python
# Convert to csv
Date_State_Positive_PCR_df.to_csv(r'F:/ETL_Project/Data_for_Analysis/PositivePCR.csv', index = False, header = True) 
```
